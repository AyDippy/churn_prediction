## Data Exploration

```python
#Analysing the ages
sns.histplot(data= df, x = 'Age', color='blue', label='Blue Histogram')

#Majority of customers are between the ages of 30 and 50
```

```python
#Grouping the ages into categories
#Below are the age categories according to the American Medical Associations
#child = 0 - 12
# Adolescent = 13 - 17
# Adults = 18 - 64
# Older Adults = 65 and above
age_cate = []
for age in df['Age']:
    if 0 >= age >= 12:
        age_cate.append('Child')
    elif 13 >= age >= 17:
        age_cate.append('Adolescent')
    elif 18 <= age <= 64:
        age_cate.append('Adult')
    elif age >= 65:
        age_cate.append('Older adults')
    
```

```python
#Added a new column to the dataset called "age_category", it categorises the ages
df['age_category'] = age_cate
```

```python
#Viewing the dqtaset
df.head()
```

```python
#Visualizing the age categories column
sns.countplot(data= df, x = 'age_category')

#It can be seen that majority of customers fall between the ages of 18 and 64 categorized as Adults
```

```python
#Checking for the gender that churned the most

sns.countplot(data = df, x= 'Exited', hue= 'Gender')

#Females tend to churn more than men from the visualization
```

```python
df.head()
```

```python
#By Categorizing credit scores into;
# excellent = 800 - 850
# very good = 740 - 799
# good = 670 - 739
# fair = 580 - 669
# poor = 300 - 579

CR_categorized = []
for score in df['CreditScore']:
    if 300 <= score  <= 579:
        CR_categorized.append('poor')
    elif 580 <= score <= 669:
        CR_categorized.append('fair')
    elif 670 <= score <= 739:
        CR_categorized.append('good')
    elif 740 <= score <= 799:
        CR_categorized.append('very good')
    elif 800 <= score <= 850:
        CR_categorized.append('excellent')
```

```python
df['cr_category'] = CR_categorized
```

```python
df.head()
```

```python
#Checking for the total number of customers that churned based on each credit score category
df.groupby(['Exited', 'cr_category'])['Exited'].count()

#Much cannot be said from this, thus the need to check for percentage that churned for each category
```

```python
#Checking the total number of customers for each category of credit score
total_cr = df['cr_category'].value_counts()
```

```python
#Checking for customers that churned for each category of credit score
churned_cr = df['cr_category'][df['Exited'] == 1].value_counts()
```

```python
total_cr
```

```python
churned_cr
```

```python
cr_rating = ['poor', 'fair', 'good', 'very good', 'excellent']
```

```python
#Calculating the percentage churned to each of the credit score categories
cr_perc = []
for r in cr_rating:
    cr_perc.append(round(churned_cr[r]/total_cr[r] * 100, 2))
```

```python
cr_perc
```

```python
#Viewing the percentage churned for every of the credit score rating
for (perc,rating) in zip(cr_perc, cr_rating):
    print(f'The percentege churned for {rating} credit score rating is %{perc}')
```

```python
#Visualizing the percentage churned for each credit score rating
plt.pie(cr_perc, labels=cr_rating, autopct='%1.1f%%')
plt.title('Percentage Churned for each credit score category')
```
